```python
import pandas as pd
import numpy as np
print(pd.__version__)
```

```python
arr=np.array([1,2,3,4,5])
view = pd.Series(arr)
print(view)
```

```python
# given it a predetermined index
view =pd.Series(arr,index=['a','b','c','d','e'])
print(view)
```

```python
# using dictionary as the input
# in dictionary the key becomes the index and the value becomes the value
dic = {"one":1,"two":2,"three":3,"four":4}
view =pd.Series(dic)
print(view)
```

```python
# you can also access it using the keys
print(view["two"])
```

```python
# by using dictionary you can also select the item you want to add to the Series
view = pd.Series(dic,index=["two","one"])
print(view)
```

```python
# By using dataframe is likee a whole table
a= pd.DataFrame(dic,index=["three","one"])
print(a)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
s =pd.DataFrame(data)
print(s)
```

```python
# && adding index to the dataframe
s= pd.DataFrame(data,index=["one","two","three"])
print(s)
```

```python
# To access a particular index u use the loc part
print(s.loc["one"])
```
